```python
#Media da idade e Desvio padrao.

media_idade = df['Idade'].mean()
desvio_idade = df['Idade'].std()

comparacao_idade1 = media_idade - desvio_idade
comparacao_idade2 = media_idade + desvio_idade

print(f"A média de idade é de: {media_idade}")
print(f"O desvio padrão da idade é de: {desvio_idade:.2f} ")
print(f"Comparando o resultado da média com o desvio padrão, grande parte das idades estão entre: {comparacao_idade1:.2f} anos abaixo da média e {comparacao_idade2:.2f} anos acima da média ")
```

```python
#Media de compras e desvio padrão

media_compras = df["Número de Compras"].mean()
desvio_compras = df["Número de Compras"].std()
comparacao_compras1 = media_compras - desvio_compras
comparacao_compras2 = media_compras + desvio_compras

print(f"A média do número de compras é de: {media_compras}")
print(f"O desvio padrão do número de compras é de: {desvio_compras:.2f} ")
print(f"Comparando o resultado da média com o desvio padrão, grande parte das compras estão entre: {comparacao_compras1:.2f} abaixo da média e {comparacao_compras2:.2f} acima da média ")
```

```python
#Media de salários e desvio padrão

media_salario = df["Salário"].mean()
desvio_salario = df["Salário"].std()
comparacao_salario1 = media_salario - desvio_salario
comparacao_salario2 = media_salario + desvio_salario


print(f"A média de salario é de: R${media_salario:.2f}")
print(f"O desvio padrão do salário é de: R${desvio_salario:.2f}")
print(f"Comparando o resultado da média com o desvio padrão, grande parte dos salários estão entre: R${comparacao_salario1:.2f} abaixo da média e R${comparacao_salario2:.2f} acima da média ")
```

```python
#Nome dos clientes com maior salário

cliente_salario = df[ ["Nome","Salário"] ]
cliente_salario_alto = cliente_salario.sort_values(by='Salário', ascending=False)

cliente_salario_alto
```

```python
#Tabela contendo o numero medio de compras por cada sexo.

sexo = df.groupby('Sexo') ['Número de Compras'].mean()

print("Numero medio de compras por cada sexo:")

sexo
```

```python
#Tabela contendo o numero medio de compras por cada sexo, com idade maior de 40 anos, por idade


df_maior_40 = df[df['Idade'] > 40]


tabela_media_compras = df_maior_40.groupby(['Sexo', 'Idade']) ['Número de Compras'].mean()


tabela_media_compras
```

```python
#Tabela contendo o numero medio de compras por cada sexo, com idade maior de 40 anos, no geral.


df_maior = df[df['Idade'] > 40]


tabela_media_compras_40 = df_maior_40.groupby('Sexo') ['Número de Compras'].mean()


tabela_media_compras_40
```

```python
#Grafico com colunas contendo o número medio de compras feita por idade dos clientes

media_compras_por_idade = df.groupby('Idade')['Número de Compras'].mean().reset_index()


plt.figure(figsize=(12, 6))
x = np.arange(len(media_compras_por_idade['Idade']))

plt.bar(x, media_compras_por_idade['Número de Compras'], width=0.6)

plt.title('Número Médio de Compras por Idade')
plt.xlabel('Idade')
plt.ylabel('Número Médio de Compras')


plt.xticks(x, media_compras_por_idade['Idade'], rotation=0, ha='center')
plt.grid(axis='y', linestyle='--', alpha=0.6)


plt.tight_layout()
plt.show()
```
